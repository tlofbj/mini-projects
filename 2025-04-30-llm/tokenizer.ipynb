## Helper Function

```python
import numpy as np

def nn_predict_next_words(x, model, prompt):
    #Ensure prompt is at least 5 words
    if(len(prompt.split(" "))) < 5:
        print("input needs at least 5 words")
        return
    #Tokenize prompt
    # tokenizer.fit_on_texts([prompt]) # DEPRECATED!!!
    token_dictionary = {index: word for word, index in tokenizer.word_index.items()}
    data = pd.DataFrame(columns=["token1", "token2", "token3", "token4", "token5", "target"])

    # User model to make predictions
    predictions = []
    next_prompt = prompt
    for i in range(x):
        tokenized_prompt = tokenizer.texts_to_sequences([next_prompt])[0]
        input = pd.DataFrame({
            'token1': [tokenized_prompt[0]], 
            'token2': [tokenized_prompt[1]], 
            'token3': [tokenized_prompt[2]], 
            'token4': [tokenized_prompt[3]], 
            'token5': [tokenized_prompt[4]]
            })
        
        result = model.predict(input)[0]

        if not isinstance(result, np.integer):
            result = np.argmax(result)

        new_word = token_dictionary[result]
        predictions.append(new_word)
        next_prompt = " ".join((next_prompt + " " + new_word).split(" ")[1:])

    print(prompt  + " " + " ".join(predictions))
```

## NN Layers

```python
from tensorflow import keras
from keras import layers

numWords = len(tokenizer.word_index) + 1
model = keras.Sequential([
    layers.Dense(units=32, input_shape=[5]),
    layers.Dense(units=32, activation='relu'),
    layers.Dense(units=32, activation='relu'),
    layers.Dense(units=32, activation='relu'),
    layers.Dense(units=numWords),
    layers.Softmax()
])

model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy,
    # optimizer='adam',
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)

# history = model.fit(train, target, epochs=1000, batch_size=32)
```

## Helper Function

```python
def getEmbeddings(text, embedding_dim, word_index):
    vocab_size = len(word_index) + 1
    embedding_index = {}
    with open("glove.6B."+str(embedding_dim)+"d.txt", "r", encoding="utf8") as f:
        for line in f:
            values = line.strip().split()
            word = values[0]
            vec = np.array(values[1:], dtype="float32")
            embedding_index[word] = vec

    embedding_matrix = np.zeros((vocab_size, embedding_dim))
    for word, i in word_index.items():
        vec = embedding_index.get(word)
        if vec is not None:
            embedding_matrix[i] = vec
    return embedding_matrix
```

```python
from tensorflow import keras
from keras import layers

numWords = len(tokenizer.word_index) + 1
print("Number of Words: ", numWords)
embedding_size = 100
embedding_matrix = getEmbeddings(file_text, embedding_size, tokenizer.word_index)

model = keras.Sequential([
    layers.Embedding(
        input_dim=numWords, 
        output_dim=embedding_size, 
        weights=[embedding_matrix],
        input_length=5,
        trainable=False),
    layers.LSTM(64),
    # layers.Dense(units=32, input_shape=[5]),
    # layers.Dense(units=64, activation='relu'),
    # layers.Dense(units=256, activation='relu'),
    layers.Dense(units=numWords),
    layers.Softmax()
])

model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy,
    # optimizer='adam',
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)

history = model.fit(train, target, epochs=100, batch_size=32)
```
